```python
import os
from dotenv import load_dotenv, find_dotenv

#Load the env file
load_dotenv(find_dotenv(), override=True)

api_key = os.getenv('OPENAI_API_KEY')
base_url = os.getenv('OPENAI_ENDPOINT')
```

```python
from langchain_openai import ChatOpenAI

#initialising the openai instance

llm = ChatOpenAI(
    api_key = api_key,
    base_url = base_url,
    model = 'azure.gpt-4.1-mini'
)

output = llm.invoke("what is google")
print(output.content)
```

```python
#using message schema for the langchain llm call
from langchain.messages import (
    HumanMessage,
    AIMessage,
    SystemMessage
)

messages = [
    SystemMessage(content = "You are a helpful ai assistant and you only need to respond in one sentence"),
    HumanMessage(content = "What is Langchain- OpenAI")
]

output = llm.invoke(messages)
print(output.content)
```
