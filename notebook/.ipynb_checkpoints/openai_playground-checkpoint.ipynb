```python
from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY, base_url=OPENAI_ENDPOINT)
def ask_ai(prompt, temp=1):
    stream = client.chat.completions.create(
        model="azure.gpt-5-nano",#OPENAI_MODEL_NAME,
        messages=[
            {
                "role":"user",
                "content":prompt
            },
            {
                "role":"system",
                "content":"you are a expert of everything and give answer only in one para and max 2 sentence"
            }
        ],
        temperature=temp,
        stream=True
    )
    # return completion.choices[0].message.content

    for chunk in stream:
        if chunk.choices[0].delta.content is not None:
            print(chunk.choices[0].delta.content, end='')
```

```python
print(ask_ai("who is Raja Ram Mohan Rai"))
```

```python
from openai import OpenAI
client =  OpenAI(
    api_key = OPENAI_API_KEY,
    base_url = OPENAI_ENDPOINT
)

file ='trans.mp3'

with open(file, 'rb') as audio_file:
    transcript = client.audio.transcriptions.create(
        
        model = 'openai.whisper',
        file = audio_file
    )

    # print(transcript)
    print(transcript.text)
```

```python
from openai import OpenAI
client = OpenAI(
    api_key=OPENAI_API_KEY,
    base_url=OPENAI_ENDPOINT
)
with client.audio.speech.with_streaming_response.create(
    model='openai.gpt-4o-mini-tts',
    voice='alloy',
    input=text
) as response:
    response.stream_to_file('tts.mp3')
```
