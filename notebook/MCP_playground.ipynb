```python
from dotenv import load_dotenv, find_dotenv
import os

load_dotenv(find_dotenv('.env'), override=True)
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
OPENAI_ENDPOINT = os.getenv("OPENAI_ENDPOINT")
```

```python
from openai import OpenAI
import os

def ask_openai(input):
    client = OpenAI(
        api_key=os.getenv("OPENAI_API_KEY"),
        base_url=os.getenv("OPENAI_ENDPOINT")
    )

    response = client.responses.create(
        model='azure.gpt-4.1',
        input=input,
        tools=[
            {
                "type": "mcp",
                "server_label": "CoinGecko",
                "server_url": "https://mcp.api.coingecko.com/sse",
                "require_approval": "never"
            }
        ]
    )

    return response
```

```python
from IPython.display import Markdown
rsp = ask_openai("create a table  and get new coint list")
Markdown(rsp.output_text)
```

```python

all_tools = [tool.name for tool in rsp.output[0].tools]
called_tools = [item.name for item in rsp.output if getattr(item, 'type', None) == 'mcp_call']
print(f'All tools: {all_tools}')
print(f'Called tools: {called_tools}')
```
