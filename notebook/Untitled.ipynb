```python
class Person:
    def __init__(self, age):
        self.age=age
p = Person(10)
q= Person(20)
if hasattr(p, 'age'):
    print(q.age)
    print(p.age)
```

```python
class Person:
    name = "Hariom"

p = Person()
q= Person()

print(q.name)
print(p.name)
q.name="Singh"
print(p.name)
print(q.name)
Person.name = "Hariom Singh"
print(p.name)
print(q.name)
del q.name
print(q.name)
```
